### Introducción.

Este conjunto de datos proviene originalmente del Instituto Nacional de Diabetes y Enfermedades Digestivas y Renales. El objetivo es predecir en base a medidas diagnósticas si un paciente tiene diabetes.

Se impusieron varias restricciones a la selección de estas instancias de una base de datos más grande. En particular, todos los pacientes aquí son mujeres de, al menos, 21 años de herencia indígena pima. Se cuentan con 768 registros en la base de datos.

* Pregnancies: número de embarazos obtenidos.
* PG Concentration: concentración de glucosa en plasma a las 2 horas en una prueba de tolerancia oral a la glucosa.
* Diastolic BP: presión arterial diastólica (mm Hg).
* Tri Fold Thick: grosor del pliegue cutáneo del tríceps (mm).
* Serum Ins: insulina sérica de 2 horas (mu U/ml).
* BMI: índice de masa corporal (peso en kg/(altura en m)^2).
* DP Function: función de pedigrí de diabetes.
* Age: edad (años).
* Diabetes: variable de clase (0 o 1).


### Definición del problema.

Este reto surge a partir de una de las problemáticas de salud más alarmantes en nuestro país. Actualmente, México se encuentra en el 6° puesto a nivel mundial en casos de diabetes y las defunciones por esta enfermedad son de aproximadamente 150,000 al año.  Por lo que se tomó la decisión de realizar una investigación acerca de la diabetes en mujeres que hayan presentado embarazos y tengan una edad, de al menos, 21 años. Esto con el fin de identificar factores clave que nos ayuden a diagnosticar si un paciente es propenso a padecer diabetes.


### Recolección de datos.

La recolección de los datos fue la parte más tediosa de este proceso debido a que no se encuentra gran cantidad de bases de datos de dominio público en la web, por lo que la investigación fue exhaustiva. Después de la búsqueda, se logró obtener una base de datos del Instituto Nacional de Diabetes y Enfermedades Digestivas y Renales quienes registraron propiedades físicas de mujeres con herencia indígena pima y provenientes de Arizona en los Estados Unidos de América.


### Identificación, categorización y limpieza de los datos.

La base de datos se encontraba, desde un comienzo, limpia ya que no se encontraban registros nulos o problemas de duplicidad con las observaciones. Por otro lado, se aplicaron diversos filtros para comprobar si una persona se encontraba enferma de diabetes o no, con el fin de observar posibles factores de los cual podamos utilizar para realizar nuestro análisis con las herramientas computacionales a nuestro alcance. Además, se analizó cada una de las variables participantes en el estudio y se realizó un análisis estadístico descriptivo, cuyos resultados develaron que se presentan diversas observaciones excepcionales que cumplen con las características de registros atípicos y extremos, en algunos casos.


### Exploración y análisis de los datos.

Al concluir la limpieza y la categorización, se emplearon gráficas y diagramas de cada campo de la base de datos para comparar a la población sana con la enferma y, de esta manera, identificar patrones que resultaran diferentes entre ambas poblaciones y, además, generar hipótesis que puedan ayudar a solucionar nuestra problemática inicial, la identificación de factores determinantes de diabetes para mujeres existentes dentro de la muestra poblacional de estudio. Entre las gráficas utilizadas, se realizaron graficas de dispersión, gráficas de barras y gráficas de pastel; las cuales nos ayudaron a identificar las relaciones que pudiesen haber entre los valores de nuestro estudio. Asimismo, se produjo una tabla de correlación para ayudarnos a observar estos datos de manera organizada e interpretar si se encuentran relaciones que no sean visibles a simple vista. Por último, se aplicó Machine Learning, con el algoritmo de máquinas de vectores de soporte, para predecir si un sujeto de la muestra presenta diabetes o se encuentra sano de acuerdo con sus datos registrados.


### Interpretación de los datos.

Finalmente, obtuvimos grandes resultados con el análisis:
* Entre mayor número de embarazos haya presentado la mujer, más probabilidad hay de que pueda desarrollar prediabetes o diabetes.
* El historial diabético no influye directamente si una persona presenta diabetes o prediabetes. Existen bastantes factores que determinan la presencia de este padecimiento en algún individuo.
* Conforme avanza la edad las personas son más propensas para presentar diabetes. Conforme el tiempo transcurre, la forma de vida tiende a sedentarizarse, además las funciones metabólicas de los organismos animales se ralentizan.
* El algoritmo desarrollado mediante Machine Learning presenta una precisión del 81% de las pruebas realizadas, tal y como se muestra en la gráfica de confusión del mismo. El entrenamiento partió con el 30% de las observaciones y consiguió resultados aceptables.
